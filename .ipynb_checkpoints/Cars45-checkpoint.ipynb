```python
grades = []
car_makes = []
car_use_conds = []
prices = []
mileages = []
years = []
car_ids = []
car_locations = []
indi_car_links = []

for each in indi_car:
    car_link = each.a.attrs['href']
    
    x = each.text
    import re
    pattern = re.compile(r'\s+')
    sentence = re.sub(pattern, '|', x)
    pat = sentence.split('|')
    pat = [line.strip() for line in pat]
    
    if 'New' in pat:
        grade = pat[1]
        car_make = ' '.join(pat[2:5])
        car_use_cond = pat[5]
        price = pat[7]
        mileage = pat[9]
        year = pat[12]
        car_id = pat[15]
        car_location =' '.join(pat[16:19])
    else:
        grade = pat[1]
        car_make = ' '.join(pat[2:5])
        car_use_cond = ' '.join(pat[5:7])
        price = pat[8]
        mileage = pat[10]
        year = pat[13]
        car_id = pat[16]
        car_location =' '.join(pat[17:19])
        
    grades.append(grade) 
    car_makes.append(car_make) 
    car_use_conds.append(car_use_cond)
    prices.append(price)
    mileages.append(mileage)
    years.append(year)
    car_ids.append(car_id)
    car_locations.append(car_location)
    indi_car_links.append(car_link)
    
```
