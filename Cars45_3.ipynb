```python
import requests
```

```python
from bs4 import BeautifulSoup
```

```python
'''
Get the page from the website and return the first 500 words of text
'''
# url = 'https://buy.cars45.com/cars?filter=&year_from=1980&year_to=2019&page=2'
url = 'https://buy.cars45.com/cars&filter=&year_from=1980&year_to=2019'
response = requests.get(url)
print(response.text[:500])
```

```python
soup = BeautifulSoup(response.text, 'html.parser')
```

```python
'''
Check and find within the soup of text generated by beautifulsoup from the website and return all divs with a class=product_box
'''
indi_car = soup.find_all('div', {'class':"product_box"})
```

```python
indi_car[26].find_all('p')
```

```python
grades = []
car_models = []
car_use_conds = []
prices = []
mileages = []
years = []
car_ids = []
car_locations = []
indi_car_links = []

for each in indi_car:
    car_link = each.a.attrs['href']
    
    grade = each.a.text.strip()
    car_model = each.h4.text.strip()
    car_use_cond = each.find_all('p')[0].text.strip()
    price = each.h5.text.strip()
    mileage = each.find_all('p')[1].text.strip()
    year = each.find_all('p')[2].text.strip()
    car_id = each.find_all('p')[3].text.strip()
    try:
        car_location = each.find_all('p')[4].text.strip()
    except:
        car_location = each.find_all('p')[3].text.strip()
    
    grades.append(grade) 
    car_models.append(car_model) 
    car_use_conds.append(car_use_cond)
    prices.append(price)
    mileages.append(mileage)
    years.append(year)
    car_ids.append(car_id)
    car_locations.append(car_location)
    indi_car_links.append(car_link)
```

```python
car_ids
```

```python
car_types = []
fuel_usages = []
gear_types = []

for each in indi_car_links:
    url = each
    response = requests.get(url)
    indi_car_soup = BeautifulSoup(response.text, 'html.parser')


    indi_car = indi_car_soup.find_all('div',{'class' :"row"} )
    
    x = indi_car[2].text
    import re
    pattern = re.compile(r'\s+')
    sentence = re.sub(pattern, '|', x)
    pat = sentence.split('|')
    pat = [line.strip() for line in pat]

    car_type = pat[1]
    fuel_usage = pat[2]
    gear_type = pat[3]
    
    car_types.append(car_type)
    fuel_usages.append(fuel_usage)
    gear_types.append(gear_type)
```
