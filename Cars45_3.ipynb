```python
grades = []
car_models = []
car_use_conds = []
prices = []
mileages = []
years = []
car_ids = []
car_locations = []
indi_car_links = []


from requests import get
from bs4 import BeautifulSoup
from time import sleep
from random import randint
from time import time;start_time = time()
from IPython.core.display import clear_output
from warnings import warn


pages = [str(i) for i in range(1,76)]

Requests = 0
start_time =time()


'''
Get the page from the website and return the first 500 words of text
'''
for page in pages:

    response = get('https://buy.cars45.com/cars?filter=&year_from=1980&year_to=2019&page=' + page)
    
    # Pause the loop
    sleep(randint(2,15))
    
    # Monitor the request
    Requests += 1
    elapsed_time = time() - start_time
    print('Request:{}; Frequency: {} requests/s'.format(Requests,Requests/elapsed_time))
    clear_output(wait=True)
    
    if response.status_code != 200:
        warn('Request: {}; Status code: {}'.format(Requests, response.status_code))
    
    if Requests > 76:
        warn('Number of requests was greater than expected.')
        
    soup = BeautifulSoup(response.text, 'html.parser')

    '''
    Check and find within the soup of text generated by beautifulsoup from the website and return all divs with a class=product_box
    '''
    indi_car = soup.find_all('div', {'class':"product_box"})

    for each in indi_car:
        car_link = each.a.attrs['href']

        grade = each.a.text.strip()
        car_model = each.h4.text.strip()
        car_use_cond = each.find_all('p')[0].text.strip()
        price = each.h5.text.strip()
        mileage = each.find_all('p')[1].text.strip()
        year = each.find_all('p')[2].text.strip()
        car_id = each.find_all('p')[3].text.strip()
        try:
            car_location = each.find_all('p')[4].text.strip()
        except:
            car_location = each.find_all('p')[3].text.strip()

        grades.append(grade) 
        car_models.append(car_model) 
        car_use_conds.append(car_use_cond)
        prices.append(price)
        mileages.append(mileage)
        years.append(year)
        car_ids.append(car_id)
        car_locations.append(car_location)
        indi_car_links.append(car_link)
        
```
