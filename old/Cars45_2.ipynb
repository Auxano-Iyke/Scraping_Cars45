```python
import requests
```

```python
from bs4 import BeautifulSoup
```

```python
'''
Get the page from the website and return the first 500 words of text
'''
url = 'https://buy.cars45.com/cars?filter=&year_from=1980&year_to=2019&page=2'
response = requests.get(url)
print(response.text[:500])
```

```python
soup = BeautifulSoup(response.text, 'html.parser')
```

```python
'''
Check and find within the soup of text generated by beautifulsoup from the website and return all divs with a class=product_box
'''
indi_car = soup.find_all('div', {'class':"product_box"})
```

```python
grades = []
car_makes = []
car_use_conds = []
prices = []
mileages = []
years = []
car_ids = []
car_locations = []
indi_car_links = []

for each in indi_car:
    car_link = each.a.attrs['href']
    
    x = each.text
    import re
    pattern = re.compile(r'\s+')
    sentence = re.sub(pattern, '|', x)
    pat = sentence.split('|')
    pat = [line.strip() for line in pat]
    
   
    if 'New' in pat:
        grade = pat[1]
        car_make = ' '.join(pat[2:5])
        car_use_cond = pat[5]
        price = pat[7]
        mileage = pat[9]
        year = pat[12]
        car_id = pat[15]
        car_location =' '.join(pat[16:19])
        
    elif 'Sold' in pat:
        grade = pat[2]
        car_make = ' '.join(pat[3:6])
        car_use_cond = ' '.join(pat[5:7])
        price = pat[8]
        mileage = pat[10]
        year = pat[13]
        car_id = pat[16]
        car_location =' '.join(pat[17:20])
       
    else:
        grade = pat[1]
        car_make = ' '.join(pat[2:5])
        car_use_cond = ' '.join(pat[5:7])
        price = pat[8]
        mileage = pat[10]
        year = pat[13]
        car_id = pat[16]
        car_location =' '.join(pat[17:19])
        
    grades.append(grade) 
    car_makes.append(car_make) 
    car_use_conds.append(car_use_cond)
    prices.append(price)
    mileages.append(mileage)
    years.append(year)
    car_ids.append(car_id)
    car_locations.append(car_location)
    indi_car_links.append(car_link)
    
```

```python
import pandas as pd
```

```python
import pandas as pd

cars45 = pd.DataFrame({'grade': grades,
'car_make': car_makes,
'car_use_cond': car_use_conds,
'price': prices,
'mileage': mileages,
'year':years,
'car_id':car_ids,
'car_location':car_locations,
'indi_car_link':indi_car_links
                       
})
print(cars45.info())
cars45.head(10)
```
